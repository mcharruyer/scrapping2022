```python
def recupMP3(mot):
    url = "https://www.ldoceonline.com/dictionary/" + mot
    response = session.get(url)
    html_doc = response.content
    soup = BeautifulSoup(html_doc,'html.parser')
    souplettesBody = soup.find('body')
    finded = souplettesBody.findAll('span',{"class":"speaker brefile fas fa-volume-up hideOnAmp"})
    finded2 = souplettesBody.findAll('span',{"class":"speaker amefile fas fa-volume-up hideOnAmp"})
    for v in finded:
        mp3 = v.get_attribute_list('data-src-mp3')
        print ("british :")
        print(mp3)
    for w in finded2:
        mp31 = w.get_attribute_list('data-src-mp3')
        print ("american :")
        print(mp31)
    
```

```python
recupMP3("cat")
```
